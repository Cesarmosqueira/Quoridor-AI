<h1>Algoritmo Dijktra</h1><br>
<b><i>El algoritmo de Dijkstra permite hallar caminos m치s cortos entre un par de nodos de un grafo.</i></b><br><br>
Como explicamos en el <a href="Trabajo_Parcial_Complejidad.ipynb">informe principal</a>, este algoritmo nos permitir치 hallar el camino m치s corto entre dos puntos en el tablero, tratando a este como una <code>matriz de adyacencia</code>. <br> <p>En este caso demostraremos la utilidad de Dijkstra en nuestro juego, hallando la distancia m치s corta entre 2 puntos del grafo.</p>


```python
n = int(input())
edges = []

for i in range(n):
    edges.append([int(x) for x in input().split()])
```

```python
start, end = [int(x) for x in input("Start and end ").split()]      

def to_adj(edges, size):
    vs = [[] for _ in range(size)]
    for src, dst, w in edges:
        vs[src].append([dst, w]) ##undirected
        vs[dst].append([src, w])
    return vs

def dijkstra(vs, start, end):
    cost = [float('inf')]*len(vs)
    parents = [-1]*len(vs)
    cost[start] = 0
    for i in range(len(vs)):
        for e, w in vs[i]:
            if parents[e] == -1:
                parents[e] = i
            new_cost = cost[i]+w
            if cost[e] > new_cost:
                cost[e] = new_cost
                parents[e] = i
    return cost[end]

adj = to_adj(edges, n)
print(f"Costo minimo de {start} a {end} = ", dijkstra(adj, start, end))
```
