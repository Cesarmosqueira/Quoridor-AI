### Grafos

Esta estructura de datos es muy utilizada para resolver problemas de toda clase. Esta conciste en `v` cantidad de nodos (o vertices), que se conenctan por `e` cantidad de aristas o edges. `G = (V, E)`. Existen dos tipos de grafos, los direccionados y los no direccionados. Los direccionados, cada arista tiene una dirección única a la que puede ir, sin embargo los grafos no direccionados no cuentan con una dirección específica.

Hay dos formas de representar un grafo:
* Lista de adyacencia

* matriz de adyacencia

En la lista, cada vertice cuenta con una lista, la cual representa los otros vertices que estan conectados con el. El numero total de listas es el numero de nodos que hay en el grafo


En cambio en la matriz, se marca como 1 si un par de vertices estan conectados, y un 0 si no lo esta. El tamaño de la matriz va a ser `V x V `
$$
a_ij=
\begin{cases}
1&& (i, j) ∈ E\\
0 && (i,j) ∉ E\\
\end{cases}
$$

Por otro lado, la matriz tiene busqueda en `O(1)` lo cual minimiza drasticamente el tiempo para ver si dos nodos estan conectados o no.


### Breath First Search
El algoritmo de breadth first search permite iterar grafos de manera iremos en orden de profundidad. Es decir, empezaremos por los nodos adyacentes a la raíz '`S`' del grafo y luego analizaremos los nodos adyacentes a ellos. <br><br>

### Backtracking

Backtracking es un algoritmo de busqueda recursiva, que analiza todas las posibilidades dentro de un grafo. Una vez que llega al final de una posibilidad, o si hay alguna condición de restricción, sale y busca el siguiente camino.

Este algoritmo se puede usar en diferentes modelos y versiones del mismo. Por ejemplo para este trabajo se usará el algorito Minimax, que es una versión del backtracking. 

Este algoritmo se puede usar para resolver problemas muy comunes, como el problema de las N-reinas, resolver un sudoku, etc.

### Algoritmo Minimax 
Minimax es un algoritmo que se basa en `backtracking` para calcular todos los posibles escenarios en un espacio de busqueda. El cual en este caso serían las decisiones que los jugadores pueden tomar.

### Algoritmo de Depth First Search

Es un algoritmo el cual busca el camino mas largo posible en el grafo, y una vez realizado, vuelve a intentar el camino mas profundo posible y así hasta completarlo. Este algoritmo usa la recursividad, aunque también hay una forma iterativa, y hasta en algunos casos podria ser mas rapido que el recursivo. Es decir, empezamos por un nodo `S` y escogemos uno de los adyacentes al mismo, y aplicamos `DFS` y así sucesivamente.

### Algoritmo de Divide y Vencerás

Es un algoritmo en la cual consiste en dividir un problema hasta poder encontrar una solución eficiente. En una forma más detallada, la técnica de este algoritmo se divide en estos tres siguientes pasos:  

1.   Dividir: El Algoritmo comienza a dividir el problema en varios subproblemas.
2.   Conquistar: Los subproblemas son llamados recursivamente hasta ser solucionados.
3. Combinar: Los subproblemas ya solucionados son combinados para ya finalmente solucionar el problema general.

### Complejidad de los Algoritmos

<ul>
    <li><h3><a href="BFS.ipynb">Breath First Search:</a></h3><ul>
<li>La complejidad de <code>BFS</code> es $O(V+E)$. siendo $V$ los vertices y $E$ las aristas del grafo formado por la cuadrícula del trablero en donde tomaremos cada posición cómo un vértice. Eso nos lleva a concluir que la complejidad del <code>BFS</code> en Quoridor sería algo cercano a $O(n^2)$ siendo $n^2$ el total de cuadrados en el tablero. Sin embargo, como usaremos <code>BFS</code> para encontrar el camino más corto, el algoritmo por lo general terminará mucho antes que se hagan $n^2$ iteraciones.  </li>
</ul></li>
<li><h3><a href="Minimax Algorithm.ipynb">Minimax:</a></h3><ul>
<li>Para llevar a cabo el algoritmo de <code>minimax</code>, habrá que definir qué tan profundo queremos buscar en los posibles movimientos del jugador. Llamaremos a esta profundidad $m$ y consideraremos el total de movimientos validos que el jugador puede realizar como $b$.  Tomando en cuenta esto y el gráfico explicado en la sección de Minimax se puede apreciar que la complejidad sería $O(b^m)$. 


Una complejidad exponencial para algo tan grande como un juego de quoridor de $n$ filas y columnas no es nada eficiente. Por eso usaremos un concepto llamado <code>Alpha beta prunning</code> el cual consiste en optimizar el rango de decisiones tomando en cuenta lo que "*necesitaría*" el jugador. Es decir, se jerarquizarán los movimientos y si se encuentra un movimiento que no ayudará al jugador, se cancelará toda la rama que nacerá de ese movimiento (<code>prunar</code> un nodo). El análisis de cada nodo antes de tomar la desición <code>prunar</code> un nodo, sería $2b-1$. Lo que significa que la complejidad final sería $O(b$^$(m/2))$. Sin embargo, no siempre habrán nodos que <code>prunar</code> entonces la posibilidad de una complejidad exponencial se mantendría latente.</li> 
</ul></li>

<h2>Teorema Maestro</h2>

El Teorema Maestro es una metodología para poder determinar el tiempo de recorrido del Algoritmo "Divide y Vencerás". Anteriormente explicado, el algoritmo "Divide y Vencerás" divide el problema en varios subproblemas, cada uno de tamaño *(n/b)* y los resuelve de forma recursiva para calcular el problema. El trabajo extra realizado para el problema viene dado por *f(n)*, es decir, el tiempo para crear los subproblemas y combinarlos da como resultado el procedimiento mismo.

$T(n) = aT(\frac{n}{b}) + \theta(n^klog^pn) $

Para este caso, nosotros utilizaremos el "Divide y Vencerás" para comprobar que las barreras del juego no bloqueen cierto punto del tablero en cada jugada. Por lo que se necesita que sea eficiente y rápido. Es por ello que mediante el Teorema Maestro vamos a poder verificar el tiempo de recorrido del algoritmo dentro de cada jugada.
