### ¿De qué trata el juego?

El juego se trata de un tablero 9x9 (nosotros lo haremos de diferentes tamaños de tablero para provar la eficiencia de los algoritmos planteados), y consta de maximo 4 jugadores, cada jugador cuenta con 1 ficha, y el objetivo es llegar al otro extremo antes que tus oponentes. Cada jugador tiene 2 opciones: moverse una casilla (de forma vertical u horizontal), o poner una pared en cualquier posición.
