### Grafos

Esta estructura de datos es muy utilizada para resolver problemas de toda clase. Esta conciste en `v` cantidad de nodos (o vertices), que se conenctan por `e` cantidad de aristas o edges. `G = (V, E)`. Existen dos tipos de grafos, los direccionados y los no direccionados. Los direccionados, cada arista tiene una dirección única a la que puede ir, sin embargo los grafos no direccionados no cuentan con una dirección específica.

Hay dos formas de representar un grafo:
* Lista de adyacencia

* matriz de adyacencia

En la lista, cada vertice cuenta con una lista, la cual representa los otros vertices que estan conectados con el. El numero total de listas es el numero de nodos que hay en el grafo


En cambio en la matriz, se marca como 1 si un par de vertices estan conectados, y un 0 si no lo esta. El tamaño de la matriz va a ser `V x V `
$$
a_ij=
\begin{cases}
1&& (i, j) ∈ E\\
0 && (i,j) ∉ E\\
\end{cases}
$$

Como se puede apreciar en la figura, el grafo mostrado es un grafo no dirigido, a lo tener una flecha en las aristas que indiquen alguna dirección. La figura b es una lista de adyacencia, donde se puede ver que, por ejemplo, el nodo 1 tiene en su lista el 2 y el 5, ya que estos son sus adyacentes. De la misma forma en la figura c (matriz) en la fila 1, se puede ver como hay 0s en la misma salvo por los nodos 2 y 5.


El usar uno u otro sera depende del problema que se te presente, ya que cada uno tiene sus ventajas y desventajas. La lista por ejemplo, para grafos pequeños es ideal, ya que no ocupa mucho espacio en memoria. Por otro lado, la matriz tiene busqueda en `O(1)` lo cual minimiza drasticamente el tiempo para ver si dos nodos estan conectados o no.


## Breath First Search
El algoritmo de breadth first search permite iterar grafos de manera iremos en orden de profundidad. Es decir, empezaremos por los nodos adyacentes a la raíz '`S`' del grafo y luego analizaremos los nodos adyacentes a ellos. <br><br>
image.png
<br>Fuente: <em> Introduccion to Algorithms Third Edition</em><br>

Tal como se ve en la figura, se analizarán los nodos de el grafo nivel a nivel hasta encontrar la forma más corta de llegar de un nodo a otro. Esta estructura se verá representada en la cuadrícula del tablero de Quoridor




## Algoritmo Minimax 
Minimax es un algoritmo que se basa en `backtracking` para calcular todos los posibles escenarios en un espacio de busqueda. El cual en este caso serían las decisiones que los jugadores pueden tomar. 
image.png
<br>Fuente: <em>https://medium.com/free-code-camp/simple-chess-ai-step-by-step-1d55a9266977</em> <br>
Estas decisiones se pueden representar como un grafo del que saldrán tantas ramas como decisiones se pueda tomar. En este caso, el espacio de busqueda serían los movimientos disponibles de cada ficha.
